## b)

Let $x$ be any real number. We will first compute $F_Y(y)=P(Y\leq x)$. Since $Y=aX+b$, we know that

$$\begin{aligned}F(y) &= P(Y\leq y) \\&= P(zX+b\leq y) \\&= P(X\leq \frac{y-b}{a}) \\&= \Phi (\frac{y-b}{b})\end{aligned}$$

So $F_Y(y)=\Phi(\frac{y-b}{a})$.

Differentiating this with respect to $y$, we find

$$\begin{aligned}f_Y(y) &= \frac{1}{a}\Phi(\frac{y-b}{a})\\&=\frac{1}{a}\phi(\frac{y-b}{a})\\&=\frac{1}{\sqrt{2\pi}a}exp{-(\frac{(y-b)^2}{2a^2})}\end{aligned}$$.

The 3 most important factors that advance modern deep neural networks are:

- Increasing dataset sizes
- Increasing accuracy
- Increasing real-world impact

Thus, the real-world impact will be a hot topic I assume.

# Q 3  Trace operator, matrix derivatives

## a) $\nabla tr(AB) = B^T$

$\because tr(AB) =\sum_{k}A_{ik}B_{kj} = \sum_{k=1}^{k}A_{1k}B_{k1}+\sum_{k=1}^{k}A_{2k}B_{k2}+...+\sum_{k=1}^{k}A_{ik}B_{kj}$


$\therefore \frac{\nabla tr(AB)}{\partial A_{ij}} = B_{ji}$

$\therefore \nabla tr(AB) = B^T$

## b) $\nabla_{A^T}f(A) = (\nabla_A f(A))^T$

$\nabla_{A^T}f(A) = \begin{bmatrix}
\frac{\partial f(A)}{\partial A_{11}} \dots \frac{\partial f(A)}{\partial A_{n1}} \\
\frac{\partial f(A)}{\partial A_{12}} \ddots \frac{\partial f(A)}{A_{n2}} \\
\vdots \\
\frac{\partial f(A)}{\partial A_{1n}} \dots \frac{\partial f(A)}{\partial A_{nn}}
\end{bmatrix} \\
=(\nabla_{A^T}f(A))^T$

## c) $\nabla_A tr(ABA^TC) = CAB +C^TAB^T$

Let's have $AB=f(A)$, where $f$ is matrix-valued

$\nabla_A tr(ABA^TC) = \nabla_A tr(f(A)A^TC) \\
=\nabla_x tr(f(x)A^TC) + \nabla_x tr(f(A)x^TC) \\
=(A^TC)^T f'(x) + (\nabla_{x^T}tr(f(A)x^TC))^T \\
=C^TAB^T + (\nabla_{x^T}tr(x^TCf(A)))^T \\
=C^TAB^T + ((cf(A))^T)^T \\
=C^TAB^T+CAB$

## d) If $z=Ax$, then $\nabla_A f(z)=(\nabla_z f(z))\otimes x$

$\because z=Ax$
